```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import re
import datetime as dt
from datetime import timedelta
import datetime as di
import matplotlib.pyplot as plt

```

```python
#Importing File:
df=pd.read_csv(r'D:\Business Analytics\Python\Project Data\Youtube.csv')
```

```python
# Data Cleaning: Strip any space in columns names:
df.columns=df.columns.str.replace(" ","")
```

```python
# Adding Missing Values into columns names:
df.columns=['video_id','trending_date','title','channels','category_id','publish_time','tags','views','likes','dislikes','comment_count','comment_disabled','rating_disable','video_error']
df
```

```python
df.shape
```

```python
#Merging  new Dataframe with categorgies names:
df2=pd.read_csv(r'D:\Business Analytics\Python\Project Data\data.csv')
df2.columns=['category_id', 'Category']
df=pd.merge(df, df2, on='category_id', how='left')
```

```python
# Re-arranging Column Names:
df=df[['video_id','title','channels','category_id','Category','trending_date','publish_time','tags','views','likes','dislikes','comment_count','comment_disabled','rating_disable','video_error']]
df
```

```python
df.publish_time.head(5)
```

```python
# Wat days of week videos are  being watched:
# Split column in to week days.
df["publishing_day"] = df["publish_time"].apply(lambda x: di.datetime.strptime(x[:10], "%Y-%m-%d").date().strftime('%a'))
count = df["publishing_day"].value_counts()
sns.barplot(x='publishing_day', y='Views Per Minute', data = df,palette='Blues',order=['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
```

```python
#String Manipulation: Made new Columns, by spliting time columnsinto time hours:
df['time']=df['publish_time'].str.split('.',expand=True).iloc[:,0]
df['time_hours']=df['time'].str.split('T',expand=True).iloc[:,1]
df['hour']=df['time_hours'].str.split(':',expand=True).iloc[:,0]
df.drop(['publish_time'],axis=1)
```

```python
#3)Percentage of viewer like and dislike with respect to publish time?
df['Percentage_of_dislike']=(df['dislikes']/df['views'])*100
df['Percentage_of_like']=(df['likes']/df['views'])*100
df.groupby('Category')['Percentage_of_like'].mean().sort_values(ascending=False)
```

```python
# Top two categories for Each Category:
Category=df[['publishing_day','Category','views']].iloc[:16,:]
def sorter(value):
    ans = value.sort_values(by='views', ascending=False)
    return ans[:2]
Category.groupby('Category').apply(sorter)
```

```python
Category=df[['publishing_day','Category','Percentage_of_like']].iloc[:,:]
def sorter(value):
    ans = value.sort_values(by='Percentage_of_like', ascending=False)
    return ans[:2]
Category.groupby('publishing_day').apply(sorter)
```

```python
#striping Time Col.
df['time'] = df['time'].str.replace("T"," ", 1)
first = pd.Series(df['time'])
date1 = di.datetime.strptime(first[12443], "%Y-%m-%d %H:%M:%S")
date1
```

```python
#Day data was published to kaggle to serve as cutoff date
date2 = di.datetime.strptime("2018-06-14 4:05:12", "%Y-%m-%d %H:%M:%S")
date2
```

```python
# Subtract the date --------------------  Trending date=date1?   Publish Date=2
elapsedtime = date2 - date1
elapsedtime
totalelapsed = (date2-date1).total_seconds()
totalelapsed/60
```

```python
def sor(value):
    ans = value.sort_values(by='Views Per Minute', ascending=False)
    return ans[:2]
df['Views Per Minute'] = df['views']/totalelapsed
df.groupby('Category').apply(sor)
```

```python
#Category Views per Minute:
sns.catplot(x='Category',y='Views Per Minute',data=df,hue='Category').fig.set_figwidth(23)
```

```python
# Plotting likes vs time( i.e likes per minutes on Youtube )
# to Make y values large , multiplied it with 10000

def sor(value):
    ans = value.sort_values(by='Views Per Minute', ascending=False)
    return ans[:2]
df['likes Per Minute'] = (df['likes']/totalelapsed)
df.groupby('Category').apply(sor)

plt.figure(figsize=(25,5))
sns.lineplot(x='Category',y='likes Per Minute',data=df)#.fig.set_figwidth(25)                #----> make wide plots 
```
